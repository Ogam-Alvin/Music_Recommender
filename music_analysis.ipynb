```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Load and view the dataset
tracks = pd.read_csv('dataset.csv', encoding = 'utf-8')
tracks
```

```python
#look at the info for tracks
tracks.info()
```

```python
#get a description of tracks
tracks.describe()
```

```python

for column in tracks.columns:
    unique_count = tracks[column].nunique()
    print(f"In '{column}' has {unique_count} unique values.")
```

```python
    #Check for missing values in the dataframe
tracks.isnull().sum()
```

```python
#remove the msiing rows since they are of no value/ will cause no harm
tracks = tracks.dropna()
#Ensure all missing data has been removed
tracks.isnull().sum()
```

```python
#check for any duplications within tracks
tracks.duplicated().sum()
```

```python
#rename the Unnamed 0: columnto user_id for easier use
tracks.rename(columns={'Unnamed: 0' : 'user_ids'}, inplace=True)
#set the count to start from 1
tracks['user_ids'] = pd.factorize(tracks['user_ids'])[0]+1
tracks.head(10)
```

```python
#Give the track_id a numerical systematic format
tracks['track_id_numeric'] = pd.factorize(tracks['track_id'])[0]+1
tracks.head(10)
```

```python
# check the datatypes for the columns
tracks.dtypes
```

```python
#Standardize and normalize where neccessary
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
numerical_cols = ['danceability', 'energy', 'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']
tracks[numerical_cols] = scaler.fit_transform(tracks[numerical_cols])
```

```python
plt.figure(figsize=(46,25))
sns.barplot(x = 'track_genre', y = 'popularity', data = tracks, palette="Blues_d")

plt.xlabel('Genre')
plt.xticks(rotation = 45)
plt.ylabel('Popularity')
plt.title( 'Genre Popularity', fontsize = 50)
#plt.savefig('Genres vs Popularity')
plt.tight_layout()
plt.show()
```

```python
top_genres = tracks.groupby('album_name')['popularity'].mean().sort_index(ascending=False).head(100)
plt.figure(figsize=(50,25))
sns.barplot(x=top_genres.index, y=top_genres.values)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.title('Top 100 Albums', fontsize = 50)
plt.xlabel('Genre')
plt.xticks(rotation = 45)
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

### KNN MODEL

The KNN ( K-Nearest Neighbors) model will be used as the baseline model for our case. The concept behind this model is that it will choose songs based on a user's past choice os songs to listen to based on the attributes e.g. loudness, echo, speechiness e.t.c. 

```python
# start by ensuring that the numerical columns are scaled

features = tracks_exploded[numerical_cols]
#check whether they have been scaled

scale_check = features.agg(['mean', 'std']).T
scale_check
```

From the check above it is clear to see that the data has been scaled and is ready to be used. This is because it is evident that the means of the features are all close to 0 and the standard deviations are all approximately 1

Now let's create the KNN model

```python
from sklearn.neighbors import NearestNeighbors

k = 5
knn = NearestNeighbors(n_neighbors=k, metric='cosine')
knn.fit(features)
```

```python
features.index
```

```python
tracks_exploded.isnull().sum()
```

```python
#check for reccomendations
song = 25 #choose the song track
distances, indices = knn.kneighbors([features.loc[song]])
recommended_Songs = tracks_exploded.iloc[indices[0]]

#display the results
print("Recommended Songs Based on:", tracks_exploded.iloc[song]['track_name'])
print(recommended_Songs[['track_name', 'artists', 'album_name']])
```
